```python
from google.colab import files
uploaded = files.upload()  # This opens a file chooser

import pandas as pd

# Get the filename of the uploaded file
filename = list(uploaded.keys())[0]

# Read the CSV file
df = pd.read_csv(filename)

# Confirm successful load
print("✅ Dataset loaded successfully!")
print("Shape:", df.shape)
df.head()
```

```python
df.columns.tolist()
```

```python
print("Initial shape:", df.shape)
print("\nMissing values per column:\n", df.isnull().sum())
```

```python
before = df.shape[0]
df.drop_duplicates(inplace=True)
after = df.shape[0]
duplicates_removed = before - after
```

```python
# There may not be many nulls, but we’ll still handle them
df.fillna({
    'Gender': 'unknown',
    'Neighbourhood': 'unknown'
}, inplace=True)
```

```python
df['Gender'] = df['Gender'].str.strip().str.upper().replace({
    'F': 'FEMALE',
    'M': 'MALE'
})
df['Neighbourhood'] = df['Neighbourhood'].str.strip().str.title()
```

```python
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'], errors='coerce')
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'], errors='coerce')
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce').astype('Int64')
df['Scholarship'] = df['Scholarship'].astype(int)
df['Hipertension'] = df['Hipertension'].astype(int)
df['Diabetes'] = df['Diabetes'].astype(int)
df['Alcoholism'] = df['Alcoholism'].astype(int)
df['Handcap'] = df['Handcap'].astype(int)
df['SMS_received'] = df['SMS_received'].astype(int)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# Standardize 'no-show' column: Yes/No instead of Y/N or similar
df['no-show'] = df['no-show'].str.strip().str.upper().replace({
    'NO': 'SHOWED',
    'YES': 'NO_SHOW'
})
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```

```python
summary = {
    "Rows after cleaning": df.shape[0],
    "Columns": df.shape[1],
    "Duplicates removed": duplicates_removed,
    "Missing values remaining": int(df.isnull().sum().sum()),
    "Date columns standardized": ["ScheduledDay", "AppointmentDay"],
    "Renamed columns": list(df.columns)
}

print("\n=== 🧼 CLEANING SUMMARY ===")
for k, v in summary.items():
    print(f"{k}: {v}")
```
